Defect localization: Highlight regions where defects are present

## 2. Mathematical Formulation

Let the input image be:

$$X \in \mathbb{R}^{H \times W \times C}$$

where:

*   $H$: height of the image
*   $W$: width of the image
*   $C$: number of channels (e.g., 3 for RGB)

The output is defined as:

**Classification output:**

$$y \in \{0, 1\}$$

where:

*   $0$: Non-defective
*   $1$: Defective

**Optional localization output:**

$$M \in \mathbb{R}^{H \times W}$$

where $M$ represents a pixel-wise defect mask.( M is an image sized map that shows where the defect is )

The learning objective is to approximate a function:

$$f(X) \to y$$  

or

$$f(X) \to (y, M)$$  Model gives 2 outputs

y → Is it defective?
M → Where is the defect?

# Create a Binary Classification Dataset
A CNN classifier cannot properly learn defective vs non-defective if training has only one class.

1. Create new folders

YouShould make this new structure:


````python
#CREATE BINARY DATASET
import os
import shutil
from sklearn.model_selection import train_test_split

original_path = "/content/bottle"
binary_path = "/content/bottle_binary"

# Create folders
folders = [
    "train/good",
    "train/defective",
    "test/good",
    "test/defective"
]

for folder in folders:
    os.makedirs(os.path.join(binary_path, folder), exist_ok=True)

# Copy all train good images
src_good_train = os.path.join(original_path, "train", "good")
dst_good_train = os.path.join(binary_path, "train", "good")

for file in os.listdir(src_good_train):
    shutil.copy(os.path.join(src_good_train, file), os.path.join(dst_good_train, file))

# Collect all defective images
defect_folders = ["broken_large", "broken_small", "contamination"]
all_defect_images = []

for defect in defect_folders:
    defect_path = os.path.join(original_path, "test", defect)
    for file in os.listdir(defect_path):
        all_defect_images.append(os.path.join(defect_path, file))

# Split defective images into train and test
train_defects, test_defects = train_test_split(all_defect_images, test_size=0.3, random_state=42)

# Copy train defective images
for file_path in train_defects:
    shutil.copy(file_path, os.path.join(binary_path, "train", "defective", os.path.basename(file_path)))

# Copy test defective images
for file_path in test_defects:
    shutil.copy(file_path, os.path.join(binary_path, "test", "defective", os.path.basename(file_path)))

# Copy test good images
src_good_test = os.path.join(original_path, "test", "good")
dst_good_test = os.path.join(binary_path, "test", "good")

for file in os.listdir(src_good_test):
    shutil.copy(os.path.join(src_good_test, file), os.path.join(dst_good_test, file))

print("Binary dataset created successfully.")
````

# STAGE 6: MODEL BUILDING AND TRANING
1. Create DataLoaders

````python
from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

print("Train batches:", len(train_loader))
print("Test batches:", len(test_loader))
````

````python
#2. Import model libraries
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models
````

````python
#3. Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
````

````python
#4. Load pretrained ResNet18
model = models.resnet18(pretrained=True)
num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 2)
model = model.to(device)
````

````python
#5. Use weighted loss
class_counts = [21, 209]  # [defective, good]
weights = torch.tensor([sum(class_counts)/class_counts[0], sum(class_counts)/class_counts[1]], dtype=torch.float32).to(device)

criterion = nn.CrossEntropyLoss(weight=weights)
optimizer = optim.Adam(model.parameters(), lr=0.001)
````

````python
#TRANING LOOP
num_epochs = 5

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    epoch_loss = running_loss / len(train_loader)
    epoch_acc = 100 * correct / total

    print(f"Epoch [{epoch+1}/{num_epochs}] - Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.2f}%")
````

````python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np

model.eval()
all_labels = []
all_preds = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        _, predicted = torch.max(outputs, 1)

        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(predicted.cpu().numpy())

print("Test Accuracy:", accuracy_score(all_labels, all_preds))
print("\nClassification Report:\n")
print(classification_report(all_labels, all_preds, target_names=test_dataset.classes))
print("Confusion Matrix:\n", confusion_matrix(all_labels, all_preds))
````

The model achieved a test accuracy of approximately 79%. The confusion matrix shows that all defective samples were correctly identified, resulting in a recall of 1.00 for the defective class. This indicates that the model is highly effective in detecting defects.

However, some normal images were misclassified as defective, leading to a lower recall for the good class. This suggests that the model is biased towards defect detection, likely due to class imbalance in the training data.

From an industrial perspective, this behavior is acceptable, as detecting all defective products is more critical than incorrectly flagging some normal products.
